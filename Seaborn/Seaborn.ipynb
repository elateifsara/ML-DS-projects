# Plot a histogram

La fonction `distplot ()` retournera une estimation de la densité du noyau (KDE) par défaut. KDE aide à lisser la distribution et constitue un moyen utile d'examiner les données. # Using a pairplot
La fonction pairplot () est généralement un moyen plus pratique d'examiner les relations par paires. Dans cet exercice, nous allons créer les mêmes résultats que PairGrid en utilisant moins de code. Ensuite, nous explorerons quelques fonctionnalités supplémentaires de la représentation par paire (). Nous allons également utiliser une palette différente et ajuster la transparence des graphiques en diagonale à l'aide du paramètre alpha.

```python
# Create a pairwise plot of the variables using a scatter plot
sns.pairplot(data=df,
        vars=["fatal_collisions", "premiums"],
        kind='scatter')

plt.show()
plt.clf()
```

```python
# Plot the same data but use a different color palette and color code by Region
sns.pairplot(data=df,
        vars=["fatal_collisions", "premiums"],
        kind='scatter',
        hue='Region',
        palette='RdBu',
        diag_kws={'alpha':.5})

plt.show()
plt.clf()
```

Un domaine de personnalisation utile consiste à définir explicitement les x_vars et les y_vars que vous souhaitez examiner. Au lieu d’examiner toutes les relations par paires, cette fonctionnalité vous permet d’examiner uniquement les interactions spécifiques pouvant présenter un intérêt.

L’autre domaine de personnalisation consiste à contrôler les types de parcelles. Dans le dernier exemple, nous inclurons la régression et le graphique kde dans le graphique en paire.

```python
# Build a pairplot with different x and y variables
sns.pairplot(data=df,
        x_vars=["fatal_collisions_speeding", "fatal_collisions_alc"],
        y_vars=['premiums', 'insurance_losses'],
        kind='scatter',
        hue='Region',
        palette='husl')

plt.show()
plt.clf()
```

```python
# plot relationships between insurance_losses and premiums
sns.pairplot(data=df,
             vars=["insurance_losses", "premiums"],
             kind='reg',
             palette='BrBG',
             diag_kind = 'kde',
             hue='Region')

plt.show()
plt.clf()
```

# Building a JointGrid and jointplot
Seaborn's JointGrid combine des tracés univariés tels que des histogrammes, des tracés de tapis et des tracés de kde avec des tracés à deux variables tels que des diagrammes de dispersion et de régression. Le processus de création de ces parcelles devrait vous être familier maintenant. Ces graphiques montrent également comment Seaborn fournit des fonctions pratiques pour combiner plusieurs graphiques ensemble.

Pour ces exercices, nous utiliserons les données de partage de vélos que nous avons examinées précédemment. Dans cet exercice, nous examinerons la relation entre les niveaux d'humidité et le total des loyers afin de déterminer s'il existe une relation intéressante que nous pourrions explorer ultérieurement.

```python
df = bicycle = pd.read_csv('datasets/bike_share.csv')
```

```python
# Build a JointGrid comparing humidity and total_rentals
sns.set_style("whitegrid")
g = sns.JointGrid(x="hum",
            y="total_rentals",
            data=df,
            xlim=(0.1, 1.0)) 

g.plot(sns.regplot, sns.distplot)

plt.show()
plt.clf()
```

```python
# Create a jointplot similar to the JointGrid 
sns.jointplot(x="hum",
        y="total_rentals",
        kind='reg',
        data=df)

plt.show()
plt.clf()
```

# Jointplots and regression
Étant donné que le graphique précédent ne montre pas de relation entre l'humidité et les montants de location, nous pouvons examiner une autre variable que nous avons examinée précédemment. Plus précisément, la relation entre temp et total_rentals.

```python
# Plot temp vs. total_rentals as a regression plot
sns.jointplot(x="temp",
         y="total_rentals",
         kind='reg',
         data=df,
         order=2,
         xlim=(0, 1))

plt.show()
plt.clf()
```

```python
# Plot a jointplot showing the residuals
sns.jointplot(x="temp",
        y="total_rentals",
        kind='resid',
        data=df,
        order=2)

plt.show()
plt.clf()
```

Sur la base du graphique résiduel et de la valeur pearson, il existe une relation positive entre la température et le total des loyers.

# Complex jointplots
Jointplot est une enveloppe de commodité entourant de nombreuses fonctions JointGrid. Cependant, il est possible de superposer certaines des parcelles JointGrid sur la trame commune standard. Dans cet exemple, nous pouvons examiner les différentes distributions pour les coureurs considérés comme occasionnels par rapport à ceux qui sont enregistrés.

```python
# Create a jointplot of temp vs. casual riders
# Include a kdeplot over the scatter plot
g = (sns.jointplot(x="temp",
             y="casual",
             kind='scatter',
             data=df,
             marginal_kws=dict(bins=10, rug=True))
    .plot_joint(sns.kdeplot))
    
plt.show()
plt.clf()
```

```python
# Replicate the above plot but only for registered riders
g = (sns.jointplot(x="temp",
             y="registered",
             kind='scatter',
             data=df,
             marginal_kws=dict(bins=10, rug=True))
    .plot_joint(sns.kdeplot))

plt.show()
plt.clf()
```
